```python
import pandas as pd

data = { "Fecha": ["2023-01-01","2023-01-02", "2023-01-03","2023-01-04"],
       "Ventas":[200,150,300,250],
        "Region":["Norte","Sur","Norte","Sur"]}
df = pd.DataFrame(data)
print(df)
```

```python
#Exploracion de datos

#imprime las primeras 5 filas
print(df.head())

#Ver el numero de columnas y filas
print(df.shape)

#Ver los tipos de datos
print(df.dtypes)



```

```python
# Lipieza de datos

#Eliminar filas con valores faltantes
df = df.dropna()

#filtrar los datos

# filtrar ventas mayores a $100

df = df[df["Ventas"] > 100]

print(df)
```

```python
#Agregacion de datos

#Agrupar por region y calvular el total de ventas
ventas_por_region = df.groupby("Region")["Ventas"].sum()

print(ventas_por_region)
```

```python
#Visualizacion de datos

#Grafico de barrras

import matplotlib.pyplot as plt

ventas_por_region.plot(kind="bar")
plt.title("Ventas por Region")
plt.xlabel("Region")
plt.ylabel("Ventas Totales")
plt.show()
```

```python
#Grafico de lineas

df.plot(x="Fecha",y="Ventas", kind="line")
plt.title("Tendencia de Ventas")
plt.xlabel("Fecha")
plt.ylabel("Ventas")
plt.show()
```

```python
#Estadisticas Basicas

#Calcular medidas estadisticas
#Media, Mediana y Moda:

media = df["Ventas"].mean()
mediana = df["Ventas"].median()
moda = df["Ventas"].mode()

print(f"Media: {media}, Mediana: {mediana}, Moda:{moda}")
```

```python
#Desviacion Estandar

desviacion = df["Ventas"].std()
print(f"Desviacion estandar:{desviacion}")
```

```python
#Analisis Exploratorio de Dattos (EDA)

#Identiicar patrones y outliers

#Identificar outliers
import seaborn as sns

sns.boxplot(x=df["Ventas"])
plt.show()

#COrrelacion entre Variables

correlacion = df.corr()
#print(correlacion)

#Heatmap de correlacion

sns.heamap(correlacion, annot=True)
plt.show()

```
